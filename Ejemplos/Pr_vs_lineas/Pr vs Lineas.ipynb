```python
def numeric_to_institucion(x):
    if x=='CHL04':
        return 1
    if x=='CHL32':
        return 2
    if x=='CHL02':
        return 3
    if x=='PER03':
        return 4
    if x=='PER05':
        return 5
    if x=='PER06':
        return 6
    if x=='CHL01' or 'CHL05' or 'CHL08' or 'CHL06':
        return 7
    if x=='CHL18' or 'CHL25' or 'CHL28' or 'CHL31':
        return 8
    if x=='Sin institucion':
        return 9
    
db2['Unidad_Negocio_num'] = db2['BUSINESS_UNIT'].apply(numeric_to_institucion)
db2
```

```python
def numeric_to_month(x):
    if x=='Enero':
        return 1
    if x=='Febrero':
        return 2
    if x=='Marzo':
        return 3
    if x=='Abril':
        return 4
    if x=='Mayo':
        return 5
    if x=='Junio':
        return 6
    if x=='Julio':
        return 7
    if x=='Agosto':
        return 8
    if x=='Septiembre':
        return 9    
    if x=='Octubre':
        return 10
    if x=='Noviembre':
        return 11   
    if x=='Diciembre':
        return 12   
    
db2['Mes_numerico'] = db2['Mes de requerimiento'].apply(numeric_to_month)
db2
```
